```python
!pip install googlemaps
import googlemaps
import csv
import time

def fetch_all_waste_facilities(api_key, latitude, longitude, radius):
    gmaps = googlemaps.Client(key=api_key)
    location = (latitude, longitude)


    results = gmaps.places_nearby(location=location, radius=radius, keyword='tourist')
    facilities = results.get('results', [])
    next_page_token = results.get('next_page_token', None)


    while next_page_token:
        time.sleep(2)
        next_results = gmaps.places_nearby(page_token=next_page_token)
        facilities.extend(next_results.get('results', []))
        next_page_token = next_results.get('next_page_token', None)


    csv_filename = "tourist.csv"
    with open(csv_filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Name', 'Address', 'Latitude', 'Longitude'])

        for place in facilities:
            name = place['name']
            location = place['geometry']['location']
            address = place.get('formatted_address', '')
            writer.writerow([name, address, location['lat'], location['lng']])
        print(f"Data written to {csv_filename}")


api_key = 'AIzaSyAfXf5nHReHqRLbjuHtB2uC3oNE1l6CEVo'
latitude = 51.0816
longitude = 0.0291
radius = 100000

fetch_all_waste_facilities(api_key, latitude, longitude, radius)

```
